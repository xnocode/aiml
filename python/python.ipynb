### Nesting

```python
age = 20
has_id = True
if age >= 18:
    if has_id:
        print("You can vote")
    else:
        print("You need an ID")
else:
    print("You are underage")
```

### Match Case

```python
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day")
```
