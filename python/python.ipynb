Logical Operator


```python
a = True
b = False
print(not a)
print(a and b)
print(a or b)
```

Operator Precedence

```python
result = 10 + 5 * 2
print(result)

result = (10 + 5) * 2
print(result)

result = 2 ** 3 * 2
print(result)

result = True or False and False
print(result)
```
